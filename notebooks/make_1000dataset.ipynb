Based on the csv create the audio 1000dataset

```python
sampled_df= pd.read_csv('data/raw/1000dataset.csv')
# Define source and destination directories
source_dir = 'audio'
destination_dir = '1000dataset'
```

```python
# Function to copy files
def copy_mp3_files(df, src, dest, track_id_col='track_id', file_extension='.mp3'):
    """
    Copies mp3 files from source to destination based on track_id.

    Parameters:
    - df (pd.DataFrame): DataFrame containing track IDs.
    - src (str): Source directory path.
    - dest (str): Destination directory path.
    - track_id_col (str): Column name for track IDs in the DataFrame.
    - file_extension (str): Extension of the mp3 files.
    """
    copied_files = 0
    missing_files = []

    for track_id in df[track_id_col]:
        # Construct the full filename
        # If track_id already includes the extension, adjust accordingly
        if track_id.lower().endswith('.mp3'):
            filename = track_id
        else:
            filename = f"{track_id}{file_extension}"

        src_file = os.path.join(src, filename)
        dest_file = os.path.join(dest, filename)

        try:
            shutil.copy2(src_file, dest_file)
            copied_files += 1
        except FileNotFoundError:
            missing_files.append(filename)
        except Exception as e:
            print(f"Error copying {filename}: {e}")

    print(f"Total files copied: {copied_files}")
    if missing_files:
        print(f"Total files missing: {len(missing_files)}")
        print("Missing files:")
        for file in missing_files:
            print(f"- {file}")

# Call the function to perform the copy operation
copy_mp3_files(sampled_df, source_dir, destination_dir, track_id_col='track_id', file_extension='.mp3')
```
